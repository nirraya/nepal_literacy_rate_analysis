```python
#importing the necessary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the csv file
data=pd.read_csv('province-wise.csv')
data.head()
```

```python
#checking if any of the fields are empty
print(data.isnull().sum())
```

```python
data['Province']
```

```python
#renaming the province names as all the provinces have names now
data['Province'].loc[0],data['Province'].loc[1],data['Province'].loc[2]='Koshi','Koshi','Koshi'
```

```python
data['Province']
```

I used ChatGPT to know how I could replace these names in a concise way, and here's what I got:

```python
# create boolean masks that identify the rows to rename
mask_madesh = data['Province'] == 'Province 2'
mask_bagmati = data['Province'] == 'Province 3'
mask_lumbini = data['Province'] == 'Province 5'
mask_sudurpashchim = data['Province'] == 'Province 7'

# set the values of the 'Province' column in the matching rows to the new names
data.loc[mask_madesh, 'Province'] = 'Madesh'
data.loc[mask_bagmati, 'Province'] = 'Bagmati'
data.loc[mask_lumbini, 'Province'] = 'Lumbini'
data.loc[mask_sudurpashchim, 'Province'] = 'Sudurpashchim'
```

```python
data['Province']
```

```python
# ploting the information of koshi province
labels = ['Male', 'Female', 'Total']
literacy_rate = [data['Literacy Rate'][0], data['Literacy Rate'][1], data['Literacy Rate'][2]]

# adding color and bar style
colors = ['blue', 'pink', 'purple']
bar_style = {'alpha': 0.85, 'width': 0.5, 'edgecolor': 'black', 'linewidth': 2}
bars = plt.bar(labels, literacy_rate, color=colors, **bar_style)

# adding the labels
plt.xlabel('Sex')
plt.ylabel('Literacy Rate')
plt.title('Literacy Rate in Koshi Province')
plt.ylim(0, 100)

# displaying the actual percentage values at the top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, yval + 2, str(round(yval, 2)) + '%', ha='center', va='bottom')

# displaying the plot
plt.show()
```
